```python
final_folder_path = "data/final"

# Use glob to find files ending with _number.txt
# The pattern matches any characters (*), followed by an underscore (_),
# one or more digits (\d+), and the .txt extension.
final_file_path = os.path.join(final_folder_path, '*_qa_pairs_cleaned_ft.json')
final_matching_files = glob.glob(final_file_path)
print(final_matching_files)
```

```python
conversations = pd.concat([
    pd.read_json(name) for name in final_matching_files
]).reset_index(drop = True)

dataset = Dataset.from_pandas(conversations)
```
